## Question 9
This exercise involves the `Auto` dataset studied in the lab. Make sure that missing values are removed from the data!

### 9a: Quantitative or Qualitative?

```python
import pandas as pd

auto = pd.read_csv("../datasets/Auto.csv")
```

```python
auto = auto.drop(auto[auto.values == "?"].index)
auto = auto.reset_index(drop=True)
auto.head()
```

Qualitative: `origin`, `name`, `year`

Quantitative: `mpg`, `cylinders`, `displacement`, `horsepower`, `weight`, `acceleration`

### 9b & c: Range, Mean, and Standard Deviation

```python
import numpy as np

auto_quant = auto[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']].astype(np.float64)
auto_quant.head()
```

```python
pd.DataFrame({'range' : auto_quant.max() - auto_quant.min(),
              'mean' : auto_quant.mean(),
              'std' : auto_quant.std()})
```

### 9d: Updated Range, Mean, and Standard Deviation (Remove 10th through 85th observations)

```python
auto_quant_d = auto_quant.copy()
auto_quant_d = auto_quant_d.drop(labels=[x for x in range(10,86)], axis=0)
auto_quant_d = auto_quant_d.reset_index(drop=True)
auto_quant_d
```

```python
pd.DataFrame({'range' : auto_quant_d.max() - auto_quant_d.min(),
              'mean' : auto_quant_d.mean(),
              'std' : auto_quant_d.std()})
```

### 9e: Time to explore the predictors graphically.

```python
pd.plotting.scatter_matrix(auto_quant)
```

- `mpg` appears correlated strongly with `weight`, `horsepower`, and `displacement`, all in non-linear in the fashion of exponential decay.

- `displacement`, `horsepower`, and `weight` seem linearly related.

- `cylinders` seem to correlate with `displacement`, `horsepower`, `weight` positively, but negatively with `mpg`.

### 10a: Load in the Boston dataset

```python
from ISLP import load_data

boston = load_data("Boston")
```

### 10b: Rows and Columns

```python
print(f"Number of Rows: {boston.shape[0]}\nNumber of Columns: {boston.shape[1]}")
```

Columns
- `crim`: per capita crime rate by town
- `zn`: proportion of residential land zoned for lots over 25,000 sq.ft
- `indus`: proportion of non-retail business acres per town
- `chas`: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
- `nox`: nitrogen oxides concentration (parts per 10 million)
- `rm`: average number of rooms per dwelling
- `age`: proportion of owner-occupied units built prior to 1940
- `dis`: weighted mean of distances to five Boston employment centres
- `rad`: index of accessibility to radial highways
- `tax`: full-value property-tax rate per $10,000
- `ptratio`: pupil-teacher ratio by town
- `lstat`: lower status of the population (percent)
- `medv`: median value of owner-occupied homes in $1000s

Rows

Each row represents each of the 506 suburbs of Boston
